```python
import os
import pandas as pd

csv_files = [v for v in os.listdir() if v[0] == '8' and v[-4:] == '.csv']

dfs = []
for csv_file in csv_files:
    df = pd.read_csv(csv_file)

    if len(df.columns) != 2:
        continue

    df.columns = ['Name', csv_file]

    dfs.append(df)

merged_df = dfs[0]
for i in range(1, len(dfs)):
    df = dfs[i]

    merged_df = merged_df.merge(df, on='Name')

merged_df['Name'] = merged_df['Name'].apply(lambda x: x.split('\\')[-1].split('.')[0])
merged_df
```
